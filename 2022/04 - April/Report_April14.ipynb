### The dominant mode changes after applying the PCA correction

```python
ex2_finalModenoPCA = mpimg.imread('./finalMode_noPCA.png')
ex2_afterMode = mpimg.imread('./modeChange_at170.png')
fig = plt.figure(figsize=(35, 20))
rows = 2
cols = 2
ax1 = fig.add_subplot(rows, cols,1)

ax1.set_title('No PCA, dominant mode at 249', size = 30)
plt.imshow(ex2_finalModenoPCA)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('mode changed after applying PCA at 80,\n this is showing the developed version of the dominant mode', size=30)
plt.imshow(ex2_afterMode)

ax3 = fig.add_subplot(rows, cols, 4)
ex2_modeChnageLast = mpimg.imread('./modeChange_at220.png')
ax3.set_title('Dominant mode after applying 2nd PCA correction', size = 30)
plt.imshow(ex2_modeChnageLast)
```

```python
ex3_resplot = mpimg.imread('../result_1/jupyter/resplot_80_90.png')
ex3_afterMode = mpimg.imread('../result_1/jupyter/pca80_90_dominant.png')
fig = plt.figure(figsize=(35, 20))
rows = 1
cols = 2
ax1 = fig.add_subplot(rows, cols,1)

ax1.set_title('PCA at 80 & 90', size = 30)
plt.imshow(ex3_resplot)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('dominant mode was not changed after the 2nd PCA', size=30)
plt.imshow(ex3_afterMode)
```
