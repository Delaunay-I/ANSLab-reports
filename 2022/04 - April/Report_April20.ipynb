```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 100
X, Y = np.loadtxt('../test/jupyter/sigmaRatio_400.dat', delimiter='\t', unpack=True)
X2, Y2 = np.loadtxt('../test/jupyter/res_260PCA.dat', delimiter='\t', unpack=True)
fig1, (ax1, ax2) =plt.subplots(1, 2)

ax1.set_yscale("log")
ax1.plot(X, Y)
ax1.set_title('Sigma ratio - ITA solver')
ax1.grid()

ax2.plot(X2, Y2)
ax2.set_yscale('log')
ax2.set_title('residual')

plt.show()
```

### Using the Krylov solver, got $\sigma$ ratio for 400 steps
The $\sigma$ ratio is for the orange line residual

```python
X, Y =np.loadtxt('../result_ex2/jupyter/res_200.dat', delimiter='\t', unpack=True)
X2, Y2 = np.loadtxt('../result_ex2/jupyter/res_400.dat', delimiter='\t', unpack=True)
X3, Y3 = np.loadtxt('../result_ex2/jupyter/sigmaRatio_400.dat', delimiter='\t', unpack=True)

fig1, ax1 =plt.subplots()
fig2, ax2 = plt.subplots()

#ax1 = fig.add_subplot(2,1,1, adjustable='box', aspect=0.3)
#ax2 = fig.add_subplot(2,1,2, adjustable='box', aspect=0.3)
ax1.set_yscale('log')
ax1.plot(X, Y, label='dt = 0.0005')
ax1.plot(X2, Y2, label='dt = 0.00005')
ax1.set_title('Residual for different time steps')
ax1.grid()
ax1.legend()

ax2.plot(X3, Y3)
ax2.set_title('Sigma ratio')
ax2.grid()

# Combine all the operations and display
plt.show()
```
