```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd

plt.rcParams['font.size'] = 7
```

```python
fig = plt.figure(figsize=(12,3), dpi=130)

rows = 1; cols = 3

fig.add_subplot(rows, cols, 1)
res = pd.read_csv("./resplots/Poisson_pca_res.dat", delimiter='\t', names = ['iter', 'residual norm'])
plt.plot(res['iter'], res['residual norm'], label='PCA update')
plt.legend(fontsize='large')
plt.yscale('log')
plt.grid(linestyle = '--')

fig.add_subplot(rows, cols, 2)
res = pd.read_csv("./resplots/Poisson_matTrans_res.dat", delimiter='\t', names = ['iter', 'residual norm'])
plt.plot(res['iter'], res['residual norm'], label='Matrix Transformation')
plt.legend(fontsize='large')
plt.yscale('log')
plt.grid(linestyle = '--')

fig.add_subplot(rows, cols, 3)
res = pd.read_csv("./resplots/Poisson_DMD.dat", delimiter='\t', names = ['iter', 'residual norm'])
plt.plot(res['iter'], res['residual norm'], label='DMD - All modes')
plt.legend(fontsize='large')
plt.yscale('log')
plt.grid(linestyle = '--')
plt.show()
```

Matrix Transformation has the same effect as eliminating all DMD modes at a single update.\
In the example above, 4 modes were computed and eliminated.

1. We should be using **more than just 10 snapshots** for constructing my data: so that our analysis can **see the converging behaviour of the data**
2. Also we should provide **an even number for our r-rank svd approximation**: this way we do eigendecomposition on a 2x2 matrix, so we get a conjugate pair of eigenvalues\
**for odd truncation**, DMD finds 1 conguate pair and 1 real eigenvalue

```python
fig = plt.figure(figsize=(12,3), dpi=150)

rows = 1; cols = 3
fig.add_subplot(rows, cols, 1)
res = pd.read_csv("./resplots/Newton_ncv150_itr200_r2.dat", delimiter='\t', names = ['iter', 'residual norm'])
plt.plot(res['iter'], res['residual norm'])
plt.title('snapshots: 150, iter: 200, rank:2')
plt.yscale('log')
plt.grid(linestyle = '--')

fig.add_subplot(rows, cols, 2)
res = pd.read_csv("./resplots/Newton_ncv100_it150_r2.dat", delimiter='\t', names = ['iter', 'residual norm'])
plt.plot(res['iter'], res['residual norm'])
plt.title('snapshots: 100, iter: 150, rank:2')
plt.yscale('log')
plt.grid(linestyle = '--')

fig.add_subplot(rows, cols, 3)
res = pd.read_csv("./resplots/Newton_noUpdate.dat", delimiter='\t', names = ['iter', 'residual norm'])
plt.plot(res['iter'], res['residual norm'])
plt.title('No updates applied')
plt.yscale('log')
plt.grid(linestyle = '--')


plt.show()
```
