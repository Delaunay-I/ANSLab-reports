Unlike PCA, DMD associates every mode with a time dynamics.\
In our previous model we did linear regression on the data of V matrix, to find the slope of the converging mode.\
In DMD, we get the eigenvalues related to our modes, so there is no need to do linear regression and find the slope (eigenvalue related to our mode of interest).

Regarding the 3rd mode, the slope of the convergence rate is not equal to 3rd mode's eigenvalue. But for the other 2 modes the slope and eigenvalues match. (they are not exactly the same, have error magnitude of e-5)\
For the 3rd mode, the error has order of e-3.


| | 1st mode | 2nd mode | 3rd mode |
| ---| --- | --- | --- |
| Eigenvalues | 0.9961 | 0.9843 | 0.9654 |
| residual plot slope| 0.9962 | 0.9843 | 0.9623 |

In DMD time dynamics, they might not have the same sign. This is related to the spatial mode corresponding to the time dynamics.\
Each time dynamic has the sign that is needed to eliminate that mode in the solution.\
i.e. Time dynamics show how the DMD modes are evolving in time. So, in our case, it is telling us how those modes are decaying and their signs are relative to this definition.
