```python
res = pd.read_csv("Residual.dat", delimiter='\t', names = ['iter', 'residual norm'])
```

```python
fig = plt.figure(figsize=(10, 8), dpi=80)
plt.plot(res['iter'], res['residual norm'], label='BE - \u0394t=10')
plt.legend(fontsize='x-large')
plt.yscale('log')
plt.grid(linestyle = '--')
#plt.show()
plt.savefig("Newtonian2D_BE_dt10_PCA_success.png")
```

This is for Backward Euler with time step of $\Delta t = 10$. At the time of applying the PCA update, we have log-linear convergence for all the variables.\
Something happende in this case, and after the peak in residual, the oscillations in convergnce fade away.\
The sigma ratio for this PCA update was $\sigma=5169.7$

For the case where I applied PCA update at $iter: 400$, the value for sigma was $\sigma=67.32$ and the update was not useful.
