## inner product of modes and solution
- Normalized solution and modes by their norms
$$
    \hat{\Phi}_i = \frac{\Phi_i}{\| \Phi_i \|} \quad i=1,2,\cdots,k
$$
$$
    \hat{Soln} = \frac{Soln}{\|Soln\|}
$$

- By taking the inner product of these normalized vectors we can see how much of a mode is in the solution before and after a DMD analysis
$$
    \left\langle \hat{\Phi}_i, \hat{Soln} \right\rangle \quad i=1,2,\cdots,k
$$

## inner product of updates before, at, after DMD

- normalized updates by their norms (as above)
$$
    \hat{U} = \frac{U}{\|U\|}
$$
