```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
res_normal = np.loadtxt('res_normal.dat')[:, 1]
res_lin = np.loadtxt('res_linearReg.dat')[:, 1]
res_rob = np.loadtxt('res_robustReg.dat')[:, 1]

figure = plt.figure(figsize=(10,5), dpi=160)

plt.semilogy(res_normal, label="no dmd")
plt.semilogy(res_lin, label="linear reg")
plt.semilogy(res_rob, label="robust reg")

plt.grid(which='both', alpha=0.5)
plt.yscale('log')
plt.xlabel('Index')
plt.ylabel('Value')

plt.legend()
plt.show()
```
